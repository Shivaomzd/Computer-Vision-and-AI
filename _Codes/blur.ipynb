```python
#importing all required packages
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

#import the image. It's needed to write the exact location of the image and based on my experience, use Capital letter while naming your image.
input_image = cv2.imread('H:\AI\Hand.jpg')
#change the color from BGR to RGB for opencv converts image to BGR when loading it
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
#Showing the image
plt.imshow(input_image)
plt.title("Original image")
plt.show()
print('_____________________________________________')

#Defining kernel size and anchor point. For more information, see line 37
Ksize = [5, 33 , 66 ]
Asize = [-1 , 5 , 10]

#Creating a for loop for applying different kernel sizes and anchor points to be able to track effect of these parameters.
for k in Ksize:
    for a in Asize:
        #we can measure the time to see the effect of each kernel size and anchor point on the process.
        start_time = cv2.getTickCount()
        #Applying blur method. for ore information, see line 37.
        blur = cv2.blur(input_image,(k,k) , (a,a))
        end_time = cv2.getTickCount()
        time = (end_time - start_time)/ cv2.getTickFrequency()
        #Showing the output image
        plt.imshow(blur)
        plt.title(f"blured image Kernel size={k} , Anchor point=({a},{a})")
        plt.show()
        print('time =', time)
    print('_____________________________________________')
  
    
#The method format is : cv2.blur(input_image, ksize, anchor, borderType)
#This filter is the normalized boxfilter operator wich is a spatial domain linear image filter. Also, this box filter is a low-pass filter.
#It’s operations are similar to average filtering technique. image sample and the filter kernel are multiplied to get the filtering result. 
#The filter kernel is like a description of how the filtering is going to happen, it actually defines the type of filtering.   
#Therefore we are working with a kernel (masking matrix) whose size shouldn't be smaller than 3*3. 
#anchor point is the point under which, the process of convolution of kernel and image happens.
#When -1, it means the anchor point is set on the center of the kernel.
```
