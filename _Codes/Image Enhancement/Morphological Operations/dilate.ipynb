```python
#importing all required packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

#import the image. It's needed to write the exact location of the image and based on my experience, use Capital letter while naming your image.
input_image = cv2.imread('H:\AI\Test.jpg')
#change the color from BGR to RGB for opencv converts image to BGR when loading it
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
#Showing the image
plt.imshow(input_image)
plt.title("Original image")
plt.show()
print('_____________________________________________')

#Defining different structuring element sizes, and iterations to study their effect on the dilation process.
#The shape mode of structuring element is cv2.MORPH_RECT which is 0 and anchor point is (-1,-1). For more information, see getStructuringElement code
kernel_sizes = [(15,15) , (20,30)]
iterations =[1,150]

#Creating a for loop for applying different iterations and kernel sizes
for kernel_size in kernel_sizes:
    for iteration in iterations:
        structuring_element = cv2.getStructuringElement(0, kernel_size, (-1,-1))
        #Applying dilate method. for ore information, see line 73
        dilated_image = cv2.dilate(input_image, structuring_element , iteration)
        #Showing the output image
        plt.imshow(dilated_image)
        plt.title(f"Dilated Image, Kernel Size= {kernel_size}, Shape Method =0, Anchor point=(-1,-1), Iteration={iteration} ")
        plt.show()
    print('_____________________________________________') 
print('*******************************************************')

#Defining different possible shapes for structuring element (mask)
#cv2.MORPH_RECT=0, #cv2.MORPH_CROSS=1, #cv2.MORPH_ELLIPSE=2


#Creating a structuring_element for applying cv2.MORPH_RECT while iterations, kernel sizes and anchor points are constant
structuring_element = cv2.getStructuringElement(0, (40,40), (-1,-1))
#Applying dilate method
dilated_image_s0 = cv2.dilate(input_image, structuring_element ,1)

#Creating a structuring_element for applying cv2.MORPH_CROSS while iterations, kernel sizes and anchor points are constant
structuring_element = cv2.getStructuringElement(1, (40,40), (-1,-1))
#Applying dilate method
dilated_image_s1 = cv2.dilate(input_image, structuring_element ,1)


#Creating a structuring_element for applying cv2.MORPH_ELLIPSE while iterations, kernel sizes and anchor points are constant
structuring_element = cv2.getStructuringElement(2, (40,40), (-1,-1))
#Applying dilate method
dilated_image_s2 = cv2.dilate(input_image, structuring_element ,1)


    
#displaying output images in a 3 in 1 subplot
images = [dilated_image_s0 ,dilated_image_s1, dilated_image_s2]
title =  ['cv2.MORPH_RECT','cv2.MORPH_CROSS','cv2.MORPH_ELLIPSE']
print(f'Dilated image Kernel size=(40,40), Anchor point=(-1,-1), Iteration= 1')
for i in range(3):
    plt.figure(figsize=(7,9), dpi=80)
    plt.subplot(3, 1, i+1)
    plt.title(title[i])
    plt.imshow(images[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()

    
    
    
#The method format is : cv2.dilate(input_image, output_image, kernel ,anchor, iterations, borderType, borderValue)
#structuring element used for dilation; if elemenat=Mat(), a 3 x 3 rectangular structuring element is used. It can be defined in the code or can be created 
#with getStructuringElement method. for more information, see getStructuringElement code.
#position of the anchor within the element; default value (-1, -1) means that the anchor is at the element center.
#If use getStructuringElement method, it will be defined in the mentioned method.
#iterations is the number of times dilation is applied.
#borderType is pixel extrapolation method.There are seven border types.
#borderValue is border value in case of a constant border
```
