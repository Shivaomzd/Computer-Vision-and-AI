```python
#importing all required packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline



#import the image. It's needed to write the exact location of the image and based on my experience, use Capital letter while naming your image.
input_image = cv2.imread('H:\AI\Circles.png')
#change the color from BGR to RGB for opencv converts image to BGR when loading it
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
#Showing the image
plt.imshow(input_image)
plt.title("Original image")
plt.show()
print('_____________________________________________')

#Defining structuring element size, shape as rectangle and anchor point at (-1,-1) which is the center of the kernel.
#For more information, see getStructuringElement code
#Morph Types are as shown below. For each Type, the operation computes different things. See line
#cv2.MORPH_ERODE = 0,
#cv2.MORPH_DILATE = 1,
#cv2.MORPH_OPEN = 2,
#cv2.MORPH_CLOSE = 3,
#cv2.MORPH_GRADIENT = 4,
#cv2.MORPH_TOPHAT = 5,
#cv2.MORPH_BLACKHAT = 6,




#Applying MORPH_ERODE = 0
structuring_element = cv2.getStructuringElement(0,(30,30), (-1,-1))
#Applying morphologyEX method. iteration equals 1. 
output_1 = cv2.morphologyEx(input_image,0, structuring_element,1)



#Applying MORPH_DILATE = 1
structuring_element = cv2.getStructuringElement(0,(30,30), (-1,-1))
#Applying morphologyEX method. iteration equals 1. 
output_2 = cv2.morphologyEx(input_image,1, structuring_element,1)



#Applying MORPH_OPEN = 2
structuring_element = cv2.getStructuringElement(0,(30,30), (-1,-1))
#Applying morphologyEX method. iteration equals 1. for ore information, see line 99
output_3 = cv2.morphologyEx(input_image,2, structuring_element,1)



#Applying MORPH_CLOSE = 3
structuring_element = cv2.getStructuringElement(0,(30,30), (-1,-1))
#Applying morphologyEX method. iteration equals 1. for ore information, see line 100
output_4 = cv2.morphologyEx(input_image,3, structuring_element,1)



#Applying MORPH_GRADIENT = 4
structuring_element = cv2.getStructuringElement(0,(30,30), (-1,-1))
#Applying morphologyEX method. iteration equals 1. for ore information, see line 101
output_5 = cv2.morphologyEx(input_image,4, structuring_element,1)



#Applying MORPH_TOPHAT=5
structuring_element = cv2.getStructuringElement(0,(30,30), (-1,-1))
#Applying morphologyEX method. iteration equals 1. for ore information, see line 102
output_6 = cv2.morphologyEx(input_image,5, structuring_element,1)


#Applying MORPH_BLACKHAT = 6
structuring_element = cv2.getStructuringElement(0,(30,30), (-1,-1))
#Applying morphologyEX method. iteration equals 1. for ore information, see line 103
output_7 = cv2.morphologyEx(input_image,6, structuring_element,1)


#displaying output images in a 4 in 2 subplot
morphology_output=[output_1,output_2,output_3,output_4,output_5,output_6,output_7]
title =  ['MORPH_ERODE','MORPH_DILATE','MORPH_OPEN','MORPH_CLOSE','MORPH_GRADIENT','MORPH_TOPHAT','MORPH_BLACKHAT']
title_o=[]
print(f'Eroded image Kernel size=(40,40), Anchor point=(-1,-1), Iteration= 1')


fig, axs = plt.subplots(1,len(morphology_output))
fig.set_figwidth(25)
fig.set_figheight(10)
fig.suptitle(title,fontsize=16)
for i in range(len(morphology_output)):
    axs[i].imshow(morphology_output[i])
plt.show()
    
    
    
#The method format is : cv2.morphologyEx(input_image, op, kernel ,anchor, iterations, borderType, borderValue)
#op is the type of a morphological operation. 
#MORPH_OPEN-- an erosion, followed by dilation
#MORPH_CLOSE-- a dilation followed by erosion
#MORPH_GRADIENT-- the difference between the dilation and the erosion of an image
#MORPH_TOPHAT-- the difference between the source image and its opening
#MORPH_BLACKHAT-- the difference between the closing of an image and the source image 
#structuring element used for dilation; if elemenat=Mat(), a 3 x 3 rectangular structuring element is used. It can be defined in the code or can be created 
#with getStructuringElement method. for more information, see getStructuringElement code.
#position of the anchor within the element; default value (-1, -1) means that the anchor is at the element center.
#If use getStructuringElement method, it will be defined in the mentioned method.
#iterations is the number of times erosion is applied.
#borderType is pixel extrapolation method.There are seven border types.
#borderValue is border value in case of a constant border.
```
