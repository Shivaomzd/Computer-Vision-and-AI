```python
#importing all required packages
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

#import the image. It's needed to write the exact location of the image and based on my experience, use Capital letter while naming your image.
input_image = cv2.imread('H:\AI\Hand.jpg')
#change the color from BGR to RGB for opencv converts image to BGR when loading it
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
#Showing the image
plt.imshow(input_image)
plt.title("Original image")
plt.show()
print('_____________________________________________')


#Defining kernel size and anchor point. For more information, see line 37
Ksize = [5, 15, 33 ]

#Creating a for loop for applying different kernel sizes to be able to track effect of these parameters.
for k in Ksize:
    #we can measure the time to see the effect of each kernel size and anchor point on the process.
    start_time = cv2.getTickCount()
    #Applying medianBlur method. for ore information, see line 37.
    blur = cv2.medianBlur(input_image,k)
    end_time = cv2.getTickCount()
    time = (end_time - start_time)/ cv2.getTickFrequency()
    #Showing the output image
    plt.imshow(blur)
    plt.title(f"blured image Kernel size={k}")
    plt.show()
    print('time =', time)
print('_____________________________________________')



#The method format is : cv2.medianBlur(input_image, ksize)
#This filter is is a non-linear filter. It replaces each pixel values (central element of the kernel) by the median values of its neighbor pixels.. 
#The filter kernel is like a description of how the filtering is going to happen, it actually defines the type of filtering.   
#kernel (masking matrix) size has to be positive and odd.
```
