```python
#importing all required packages
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

#import the image. It's needed to write the exact location of the image and based on my experience, use Capital letter while naming your image.
input_image = cv2.imread('H:\AI\Hand.jpg')
#change the color from BGR to RGB for opencv converts image to BGR when loading it
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
#Showing the image
plt.imshow(input_image)
plt.title("Original image")
plt.show()
print('_____________________________________________')



#Applying medianBlur method. for ore information, see medianBlur code.
median_blur = cv2.medianBlur(input_image,21)

#Applying blur method. for ore information, see blur code.
blur_blur = cv2.blur(input_image,(21,21) , (-1,-1))

#Applying boxFilter method. for ore information, see boxFilter code.
box_blur = cv2.boxFilter(input_image,20,(21,21) , (-1,-1))

#Applying GaussianBlur method. for ore information, see GaussianBlur code.
gaussian_blur = cv2.GaussianBlur(input_image, (21,21),0)



images = [box_blur, blur_blur, gaussian_blur, median_blur]
title =  ['box', 'blur', 'gaussian', 'median']

#displaying output images in a 2 in 2 subplot
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.title(title[i])
    plt.imshow(images[i])
    plt.xticks([])
    plt.yticks([])
```
