```python
#importing all required packages
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

#import the image. It's needed to write the exact location of the image and based on my experience, use Capital letter while naming your image.
input_image = cv2.imread('H:\AI\Test.jpg')
#change the color from BGR to RGB for opencv converts image to BGR when loading it
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
#Showing the image
plt.imshow(input_image)
plt.title("Original image")
plt.show()
print('_____________________________________________')



#Defining kernel size. For more information, see line 70
Ksize = [1,3,5,7]

#Creating a for loop for applying different kernel sizes, while scale is 1 and delta is 0
for k in Ksize: 
    #Applying Laplacian method. for ore information, see line 70
    edged_image = cv2.Laplacian(input_image, -1,ksize=k)
    #displaying output images in a 1 in 3 subplot
    print(f'Edged image Kernel size={k}')
    plt.imshow(edged_image)
    plt.show()
    print('_____________________________________________') 

    
print('**************************************************************')    
#Applying Laplacian method while scale is changing and kernel size, delta are constant   
edged_image_3 = cv2.Laplacian(input_image, -1,ksize=3, scale=3) 
edged_image_10 = cv2.Laplacian(input_image, -1,ksize=3, scale=10)
images = [edged_image_3,edged_image_10]
title =  ['scale=3','scale=10']
print(f'Edged image Kernel size=3 , delta=0')
for i in range(2):
    plt.figure(figsize=(7,9), dpi=120)
    plt.subplot(1, 2, i+1)
    plt.title(title[i])
    plt.imshow(images[i])
    plt.show()
print('_____________________________________________') 



print('**************************************************************')    
#Applying Laplacian method while delta is changing and kernel size, scale are constant    
edged_image_50 = cv2.Laplacian(input_image, -1,ksize=3, scale=1, delta=50) 
edged_image_150 = cv2.Laplacian(input_image, -1,ksize=3, scale=1, delta=150)
images = [edged_image_50,edged_image_150]
title =  ['delta=50','delta=150']
print(f'Edged image Kernel size=3 , scale=1')
for i in range(2):
    plt.figure(figsize=(7,9), dpi=120)
    plt.subplot(1, 2, i+1)
    plt.title(title[i])
    plt.imshow(images[i])
    plt.show()
print('_____________________________________________') 







#The method format is: cv2.Laplacian(input_image, ddepth,ksize, scale, delta, borderType)
#ddepth is the output format which is determined based on the depth of the input image. When equals -1, the output's depth is the same with that of the input.
#ksize is the size of the kernel which has to be 1,3,5, or 7.
#scale is optional scale factor for the computed derivative values.
#delta optional delta value that is added to the results prior to storing them in dst.As you can see in the example the higher the delta, the brighter each pixel.
```
