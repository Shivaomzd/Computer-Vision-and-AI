```python
#importing all required packages
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

#import the image. It's needed to write the exact location of the image and based on my experience, use Capital letter while naming your image.
input_image = cv2.imread('H:\AI\Hand.jpg')
#change the color from BGR to GRAY.
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
#Showing the image
plt.imshow(input_image)
plt.title("Original image")
plt.show()
print('_____________________________________________')


#Defining kernel size. For more information, see line 45
asize = [1, 7]
thresholds1=[0,90]
thresholds2=[150,250]


#Creating a for loop for applying different Aperture size, Threshold1, and Threshold2
for a in asize:
    for threshold1 in thresholds1:
        for threshold2 in thresholds2: 
            #Applying Canny method;Once with Manhttan norm(L1) and once with Euclidean norm(L2). for more information, see line 45.
            edged_image_l1 = cv2.Canny(input_image, a ,threshold1,threshold2,L2gradient=0)
            edged_image_l2 = cv2.Canny(input_image, a ,threshold1,threshold2,L2gradient=1)
            #displaying output images in a 2 in 2 subplot
            images = [edged_image_l1,edged_image_l2]
            title =  ['L1 norm','L2 norm']
            print(f'Edged image Aperture size={a}, Threshold1={threshold1}, Threshold2={threshold2}')
            for i in range(2):
                plt.subplot(1, 2, i+1)
                plt.title(title[i])
                plt.imshow(images[i])
                plt.xticks([])
                plt.yticks([])
                plt.show()
            print('_____________________________________________') 
        
        
 
# The method's format is :cv2.Canny(input_image, aperture ,threshold1,threshold2,L2gradient=0)
#It's an edge detector which has good immunity to noise and uses 1st derivatives. 
#aperture size refers to the kernel size of the Sobel operator and has to be 1,3,5 or 7. For more information, see Sobel oprtator.
#Threshold1 and Threshold2 are in fact lower and upper threshold. If a pixel's value is smaller than lower threshold, 
#operator doesn't consider it as an edge and if it's larger than threshold2(upper one), will consider it as an edge.
#If the value is between threshold1 and threshold2, will consider it as an edge if it's connected to the threshold2.

        
```
